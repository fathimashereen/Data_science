```python



















































































































































































































































































import numpy as np
from matplotlib import pyplot as plt

```

```python
#square and cube of 10 numbers in simple graph
x=list(range(1,11))
y=[a**2 for a in x ]
z=[a**3 for a in x ]
plt.plot(x,y,marker='*',linestyle='dotted',color='red')
plt.plot(x,z,marker='+',color='black')
plt.legend(['square','cube'])
plt.grid()
```

```python
#scatter plot
name=['john','sana','raha','anu','xavier','sera','jane','aisha','ahmed','rohan']
phy=np.array([50,60,10,90,45,78,80,67,89,55])
chem=np.array([55,65,35,95,55,68,84,97,100,75])
plt.scatter(name,phy,marker='H',color='pink',edgecolors='blue',s=80)
plt.scatter(name,chem,marker='v',color='yellow',edgecolors='blue',s=80)
plt.xlabel('names')
plt.ylabel('marks')
plt.legend(['physics','chemistry'])
plt.title("Marks of physics and chemistry")
```

```python
#bar plot
name=['john','sana','raha','anu','xavier','sera','jane','aisha','ahmed','rohan']
x=[170,162,152,175,145,167,161,155,180,190]
plt.bar(name,x,color='maroon',width=0.5)
plt.xlabel('Name')
plt.ylabel('Height')
plt.title('Height of 10 persons')
```

```python
#histogram
a=np.random.normal(170,2,200)
plt.hist(a,color='green')
```

```python
#pie plot
marks=[100,60,80,90,75,70]
exp=[0.1,0,0,0,0,0]
col=['pink','red','blue','green','yellow','brown']
labels=['or','daa','dsa','sna','dsm','mad']
plt.pie(marks,explode=exp,colors=col,labels=labels)
```

```python
#sine and cosine
#sine
a=np.arange(0,20,0.2)
b=np.sin(a)
plt.plot(a,b,linestyle='dotted',color='red')
#cosine
c=np.cos(a)
plt.plot(a,c,color='green')
plt.xlabel('angle')
plt.ylabel('value')
plt.legend(['sine','cosine'])
plt.title('Sine and cosine')
```

```python
#subplot
#plot1
x=[1,2,3,4]
y=[1,4,9,16]
plt.subplot(1,3,1)
plt.plot(x,y)

#plot2
x=[1,2,3,4]
y=[3,5,7,1]
plt.subplot(1,3,2)
plt.plot(x,y)

#plot3
x=[1,2,3,4]
y=[5,6,7,8]
plt.subplot(1,3,3)
plt.plot(x,y)


```
